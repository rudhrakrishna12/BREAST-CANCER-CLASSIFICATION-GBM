```python
x,y=load_breast_cancer(return_X_y=True)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python

model=GradientBoostingClassifier() 
model.fit(x_train,y_train)
```

```python
x_pred=model.predict(x_test)
```

```python
model1=GradientBoostingClassifier(n_estimators=200,min_samples_split=20,min_samples_leaf=20,subsample=1.0)
model1.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
precision=precision_score(y_test,x_pred)
recall=recall_score(y_test,x_pred)
accuracy=accuracy_score(y_test,x_pred)
f1=f1_score(y_test,x_pred)
print(f'precision:{precision :,.2f}')
print(f'recall:{recall :,.2f}')
print(f'accuracy:{accuracy :,.2f}')
print(f'f1:{f1 :,.2f}')
```

```python
def predict_cancer(*feature):
    feature=np.array(feature).reshape(1,-1)
    predicton=model.predict(feature)[0]
    probability=model.predict_proba(feature)[0]
    label='' if predicton==0 else 'benign'
    return(
        label,
        f'malignant:{probability[0]:.4f},benign:{probability[1]:.4f}'
    )
```

```python
inputs=[gr.Number(label="mean radius"),
    gr.Number(label="mean texture"),
    gr.Number(label="mean perimeter"),
    gr.Number(label="mean area"),
    gr.Number(label="mean smoothness"),
    gr.Number(label="mean compactness"),
    gr.Number(label="mean concavity"),
    gr.Number(label="mean concave points"),
    gr.Number(label="mean symmetry"),
    gr.Number(label="mean fractal dimension"),

    gr.Number(label="radius error"),
    gr.Number(label="texture error"),
    gr.Number(label="perimeter error"),
    gr.Number(label="area error"),
    gr.Number(label="smoothness error"),
    gr.Number(label="compactness error"),
    gr.Number(label="concavity error"),
    gr.Number(label="concave points error"),
    gr.Number(label="symmetry error"),
    gr.Number(label="fractal dimension error"),

    gr.Number(label="worst radius"),
    gr.Number(label="worst texture"),
    gr.Number(label="worst perimeter"),
    gr.Number(label="worst area"),
    gr.Number(label="worst smoothness"),
    gr.Number(label="worst compactness"),
    gr.Number(label="worst concavity"),
    gr.Number(label="worst concave points"),
    gr.Number(label="worst symmetry"),
    gr.Number(label="worst fractal dimension"),
]
```

```python
interface=gr.Interface(
    fn=predict_cancer,
    inputs=inputs,
    outputs=[
        gr.Textbox(label='prediction'),
        gr.Textbox(label='probabilities')
    ],
    title='gradient boosting classifier(pickle model)',
    description='predictions using a pre-training gradient boosting model'
)
interface.launch()
```
